## Hamiltonian Evolution under any Pauli Gate

```python
qc = QuantumCircuit(3)

op = Pauli('XYZ')
U = HamiltonianGate(op,1.2)
qc.append(U,range(3))
qc.draw('mpl')
```

```python
qc = QuantumCircuit(3)

op = SparsePauliOp(['XYX','YZY'])
U = HamiltonianGate(op,1.2)
qc.append(U,range(3))
qc.draw('mpl')
```

```python
qc = QuantumCircuit(2)

init = Initialize(np.random.random(4)+1j*np.random.random(4),normalize=True)
qc.append(init,range(int(np.log2(len(init.params)))))
qc.draw('mpl')
```

```python
qc = IntegerComparator(3,3)
qc.decompose().draw('mpl')
```

```python
# Define the linear function parameters
slope = 1.0
offset = 0.0
num_qubits = 3

linear_amplitude_function = LinearAmplitudeFunction(num_qubits, slope, offset,(-1,2),(0,2))

# Having additional 1 Ancilla qubit
qc = QuantumCircuit(num_qubits+1)

# Having additional 1 Ancilla qubit
qc.append(linear_amplitude_function,range(num_qubits+1))
qc.draw('mpl')
```

```python
linear_pauli_rot = LinearPauliRotations(3)
# Having additional 1 Ancilla qubit
qc = QuantumCircuit(num_qubits+1)

# Having additional 1 Ancilla qubit
qc.append(linear_pauli_rot,range(num_qubits+1))
qc.decompose().decompose().draw('mpl')
```
