```python
import numpy as np 
from qiskit.quantum_info import Operator, Pauli
from qiskit.visualization import plot_state_city, plot_state_hinton
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
op = Operator(np.array([[0,1],[1,0]]))
```

```python
#### Conjugate of Operator ######
op.conjugate()
```

```python
###### Dot product of Oprator #####
op.dot(op.conjugate())
```

```python
###### Checking if Operator is unitary ##########
op.is_unitary()
```

```python
###### Power of Operator ####### 
op.power(3)
```

### **Way to Generate Operator using Pauli Method**

```python
#### Converting Paulis to Operator #####
op = Pauli('XY')
print(op.to_matrix())
```

```python
paulis = ['XX','-iZX','XY','iYZ','ZZ','YY']
```

```python
plt.figure(figsize=(13,25))
for i in range(len(paulis)):
    plt.subplot(len(paulis),2,2*i+1)
    sns.heatmap(Pauli(paulis[i]).to_matrix().real,cmap='rocket') #drawing Real Part

    plt.subplot(len(paulis),2,2*i+2)
    sns.heatmap(Pauli(paulis[i]).to_matrix().imag,cmap='rocket')
plt.show()
```

```python
plot_state_city(Pauli('iZX').to_matrix() +Pauli('XZ').to_matrix())
```

```python
plot_state_hinton(Pauli('iZX').to_matrix() +Pauli('XZ').to_matrix(),figsize=(12,4))
```
