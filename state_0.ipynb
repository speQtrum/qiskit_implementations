```python
from qiskit.quantum_info import random_statevector, Statevector, Operator
from qiskit.circuit.library import XGate,YGate, ZGate, UnitaryGate
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector, plot_state_paulivec
import numpy as np
import matplotlib.pyplot as plt
```

#### **Generating a Quantum State**

```python
state = Statevector(np.array([1/np.sqrt(2),1/np.sqrt(2)]))
print(state)
# state.draw('paulivec')
```

```python
# Generate a quantum state from label representation
psi = Statevector.from_label('011') # |011>
print(psi.data)
plot_bloch_multivector(psi)
```

```python
plot_state_qsphere(psi,figsize=(4,4))
```

#### **Generating Random Quantum State**

```python
rand_state = random_statevector(2)
print("The Quantum state vector is:",rand_state)
print("Sample counts if we measure state:",rand_state.sample_counts(shots=1000))
print("Probability array if we measure:",rand_state.probabilities()) 
print("Probability array if we measure:",rand_state.probabilities_dict())
print("Trace of Quantum State:",rand_state.trace())
```

```python
# Plotting Quantum State in Bloch Sphere
rand_state.draw('bloch',figsize=(4,4))
```

```python
# Plotting Quantum State in Q-Sphere
rand_state.draw('qsphere',figsize=(4,4))
```

```python
# Plotting Quantum State in Hinton Plot
rand_state.draw('hinton',figsize=(16,3))
```

```python
# # Plotting Quantum State in 3D bar plot
rand_state.draw('city',figsize=(13,5))
```

```python
rand_state.draw('paulivec',figsize=(13,4))
```

```python
# # Converting Quantum State to Density matrix |psi><psi|
rand_state.to_operator()
```

```python
# Applying an Operator on a Quantum State
M = XGate().to_matrix() # Buliding a Pauli-X Gate
op = Operator(M)        # Converting the Matrix to Operator
print("Matrix of the Operator:\n",op.data)

# Applying an Operator on a Quantum State
evolved_state = rand_state.evolve(op)
print("\nState after applying operator:",evolved_state.data)
print("Expectation value of the operator:",rand_state.expectation_value(op)) # <rand_state|op|rand_state>
evolved_state.draw('bloch',figsize=(4,4))
```
